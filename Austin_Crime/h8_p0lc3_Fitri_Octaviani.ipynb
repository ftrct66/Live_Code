## SOAL 2

> Apa saja jenis tindak kejahatan yang terjadi di kota Austin selama tahun 2016?

```python
Jenis_Kejahatan = client.query('''
SELECT DISTINCT primary_type as jenis_kejahatan
FROM bigquery-public-data.austin_crime.crime
where EXTRACT(YEAR FROM timestamp) = 2016

''').to_dataframe()
Jenis_Kejahatan
```

Jenis tindak kejahatan yang terjadi di kota Asutin ada Rape, Theft, Robbery, Agg Assault, Burglary, Auto Theft, dan Murder. Dengan berbagai kejahatan ini maka kepolisian lebih teliti lagi perihal keamanan di kota Austin.

## SOAL 3

> Berapa banyak kasus tindak kejahatan berdasarkan kasus pembersihannya (clearance status) selama tahun 2016?

```python
kasus_pembersihan = client.query('''
SELECT clearance_status, COUNT(*) AS jumlah_clear
FROM bigquery-public-data.austin_crime.crime
where EXTRACT(YEAR FROM timestamp) = 2016
GROUP BY clearance_status

''').to_dataframe()
kasus_pembersihan
```

Pada tahun 2016 ini tindak kejahatan yang paling banyak adalah dengan status not cleared dengan jumlah 29715. Hal ini polisi sangat diperlukan untuk lebih menyelesaikan kasus yang belum terselesaikan.

## SOAL 4

> Jenis kejahatan apa saja yang paling banyak belum terselesaikan kasusnya di tahun 2016?

```python
belum_terselesaikan = client.query('''
SELECT primary_type as Jenis_kejahatan, clearance_status, COUNT (unique_key) as banyak_kejahatan
FROM bigquery-public-data.austin_crime.crime
where clearance_status = 'Not cleared'
      AND EXTRACT(YEAR FROM timestamp) = 2016
GROUP BY primary_type, clearance_status
ORDER BY banyak_kejahatan desc

''').to_dataframe()
belum_terselesaikan
```

Jenis kejahatan yang paling banyak tidak dibersihkan adalah theft. Sehingga polisi perlu memerhatikan keamanan sekitar dan warga di kota austin pun harus lebih waspada.

## SOAL 5

Di bulan apa di tahun 2016 kasus pencurian (theft) sangat banyak terjadi?

```python
kasus_pencurian = client.query('''
SELECT primary_type, EXTRACT (month from (clearance_date)) as Bulan,
COUNT (*) as kasus
FROM bigquery-public-data.austin_crime.crime
WHERE
EXTRACT(YEAR FROM clearance_date) = 2016
AND primary_type = 'Theft'
GROUP BY primary_type, Bulan
ORDER BY kasus desc

''').to_dataframe()
kasus_pencurian
```

Kasus theft paling banyak terjadi adalah pada bulan ke-8 dengan total 2375.

## SOAL 6


Di distrik apa yang paling banyak terjadi kasus pencurian selama 2016?

```python
distrik = client.query('''

SELECT district,
        Count (*) as theft
FROM bigquery-public-data.austin_crime.crime
WHERE
      EXTRACT(YEAR FROM clearance_date) = 2016
      and primary_type = 'Theft'
GROUP BY district
ORDER BY theft desc

''').to_dataframe()
distrik
```

Dalam kota asutin, district yang paling banyak terjadi kasus pencurian adalah district D. Untuk mengurangi tingkat kejahatan ini maka perlu waspada.
